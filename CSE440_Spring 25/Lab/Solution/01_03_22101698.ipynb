### **Task 02 ||**

```python
nltk.download('gutenberg')
nltk.download('stopwords')
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
from wordcloud import WordCloud

novel = gutenberg.fileids()
text = gutenberg.raw(novel[0])
words = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_words = [word.lower() for word in words if word.isalnum() and word.lower() not in stop_words]

fdist = FreqDist(filtered_words)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(fdist)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### **Task 03 ||**

```python
from nltk.stem import LancasterStemmer
from nltk.corpus import gutenberg
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
from wordcloud import WordCloud

novel = gutenberg.fileids()
text = gutenberg.raw(novel[0])
words = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_words = [word.lower() for word in words if word.isalnum() and word.lower() not in stop_words]

fdist_B = FreqDist(words)
most_frequent_words_before =fdist_B.most_common(15)

stemmer = LancasterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]



fdist_A = FreqDist(stemmed_words)
most_frequent_words_after =fdist_A.most_common(15)


beforeWord = [word[0] for word in most_frequent_words_before]
beforeCount = [word[1] for word in most_frequent_words_before]


plt.bar(beforeWord, beforeCount, color="blue")
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title(f"Top 15 Most Frequent Words in '{novel[0]}' (Before Preprocessing)")
plt.xticks(rotation=45)
plt.show()


afterWord = [word[0] for word in most_frequent_words_after]
afterCount = [word[1] for word in most_frequent_words_after]


plt.bar(afterWord, afterCount, color="green")
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.title(f"Top 15 Most Frequent Words in '{novel[0]}' (After Preprocessing)")
plt.xticks(rotation=45)
plt.show()
```

```python
from nltk.util import ngrams

novel = gutenberg.fileids()
text = gutenberg.raw(novel[0])
words = word_tokenize(text)

bigrams = list(ngrams(words, 2))

fdist = FreqDist(bigrams)
most_common_bigrams = fdist.most_common(50)

print(most_common_bigrams)
```
